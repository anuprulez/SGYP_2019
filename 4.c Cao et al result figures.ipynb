```python
%pylab inline
%config InlineBackend.figure_format ='retina'

import time

import anndata
import matplotlib.colors as mcolors
import pandas as pd

```

```python
adata = anndata.read('cao_atlas.h5ad')
```

```python
Z_df = pd.read_csv('cao_vae_results\ldvae_1_Z.csv', index_col=0)
W_df = pd.read_csv('cao_vae_results\ldvae_1_W.csv', index_col=0)
```

```python
Z_df = Z_df.add_prefix('z')
W_df = W_df.add_prefix('w')
```

```python
def Z_covariance(Z):
    Zcentered = Z - Z.mean(0)
    Zscaled = Z / Z.std(0)
    ZTZ = np.cov(Zscaled.T)
    
    eigen_values, _ = np.linalg.eig(ZTZ)
    singular_values = np.sqrt(eigen_values)
    variance_explained = singular_values / singular_values.sum()

    return ZTZ, variance_explained
```

```python
_, variance_explained = Z_covariance(Z_df)
idx = np.argsort(variance_explained)[::-1]
```

```python
Z_df_ordered = pd.DataFrame(Z_df.values[:, idx]).add_prefix('z')
Z_df_ordered.index = Z_df.index
W_df_ordered = pd.DataFrame(W_df.values[:, idx]).add_prefix('w')
W_df_ordered.index = W_df.index
```

```python
W_df_ordered = adata.var.loc[W_df_ordered.index].join(W_df_ordered)
```

```python
text_shift = {
    (0, 'Pax3'): (-0.4, 0.0),
    (0, 'Gm29478'): (0., 0.2),
    (0, 'Rfx4'): (0.1, 0.2),
    (0, 'Glra1'): (0.1, -0.0),
    (0, 'Cntn4'): (0.0, 0.2),
    (0, 'Lrfn5'): (0.0, -0.1),
    (0, 'Snhg11'): (0.0, 0.1),
    
    (1, 'Lrp1b'): (0.0, -0.2),
    (1, 'Grid2'): (0.0, -0.1),
    (1, 'Gm45455'): (0., 0.1),
    (1, 'Glra1'): (0., 0.2),
    
    (2, 'Vmn2r-ps49'): (0.0, 0.1),
    (2, 'Dkk2'): (0.0, -0.1),
    (2, 'Adam12'): (0.0, 0.2),
    
    (3, 'Gm45455'): (0.0, 0.3),
    (3, 'Lama2'): (0.0, -0.1),
    (3, 'Ebf3'): (-0.2, -0.0),
    (3, 'Ebf2'): (0.0, 0.1),
    (3, 'Thsd7b'): (0.2, -0.3),
    
    (4, 'Afp'): (0.4, 0.1),
    (4, 'Ttn'): (-0.1, -0.0),
    
}
```
